```python
# Second, parse xml files to dataframe to check the xmin, xmax, ymin, ymax values

def xml_to_df(xml_file):
    xml_data = open(xml_file, 'r').read()  
    root = ET.XML(xml_data) 
    filename_list = []
    label_list = []
    box_list = []
    for i, child in enumerate(root):
        if child.tag == 'path':
            filename_list.append(child.text)
        for subchild in child:
            if subchild.tag == 'name':
                label_list.append(subchild.text)
            bbox_list = [] 
            if subchild.tag == 'bndbox':
                for subchild_ in subchild:
                    bbox_list.append(subchild_.text)
            box_list.append(bbox_list)
        
    bbox_list_0 = []
    for i in range(len(box_list)):
        if len(box_list[i])>0:
            bbox_list_0.append(box_list[i])
    
    xmin_list = []
    ymin_list = []
    xmax_list = []
    ymax_list = []
    for i in range(len(bbox_list_0)):
        label = label_list[i]
        box = bbox_list_0[i]
        xmin = int(box[0])
        ymin = int(box[1])
        xmax = int(box[2])
        ymax = int(box[3])
        xmin_list.append(xmin)
        ymin_list.append(ymin)
        xmax_list.append(xmax)
        ymax_list.append(ymax)
    
    label_ = pd.Series(label_list, name='label')
    xmin_ = pd.Series(xmin_list, name='xmin')
    ymin_ = pd.Series(ymin_list, name='ymin')
    xmax_ = pd.Series(xmax_list, name='xmax')
    ymax_ = pd.Series(ymax_list, name='ymax') 

    filename_list_ = [filename_list[0]]*len(label_list)
    filename_list_ = pd.Series(filename_list_, name='filename')
    df = pd.concat([filename_list_, label_, xmin_, ymin_, xmax_, ymax_], axis=1)
    return df
    
```

```python
df_list = []
for i in range(len(xml_list)):
    df = xml_to_df(xml_list[i])
    df_list.append(df)
```

```python
def create_xml(directory, imgfilepath, object_list, size_tuple, savedir):
    """
    params:
    - imgfilepath: path of corresponding img file. only the basename will be actually used.
    - object_list: python list of element in tuple of (classname, [x1,y1,x2,y2])
    - savedir: output directory to save generated xml file
    """
    basename = os.path.basename(imgfilepath)
    filename, _ = os.path.splitext(basename)
    root = et.Element('annotation')
    
    fn_elem = et.SubElement(root, 'filename')
    bn = os.path.basename(imgfilepath)
    fn, _ = os.path.splitext(bn)
    img_bn = f"{fn}.jpg"
    fn_elem.text = img_bn

    path = et.SubElement(root, 'path')
    if 'png' in imgfilepath:
        imgfilepath = imgfilepath[:-3] + str('jpg')
    elif 'jpg' in imgfilepath:
        imgfilepath = imgfilepath
    path.text = imgfilepath
    print('imgfilepath', imgfilepath)

    source = et.SubElement(root, 'source')
    database = et.SubElement(source, 'database')
    database.text = 'Unknown'

    size = et.SubElement(root, 'size')
    width = et.SubElement(size, 'width')
    width.text = str(size_tuple[0])
    height = et.SubElement(size, 'height')
    height.text = str(size_tuple[1])
    depth = et.SubElement(size, 'depth')
    depth.text = str(size_tuple[2])
    
    for classname, (x1,y1,x2,y2) in object_list:
        object_elem = et.SubElement(root, 'object')
        name = et.SubElement(object_elem, 'name')
        name.text = classname
        
        pose = et.SubElement(object_elem, 'pose')
        pose.text = 'Unspecified'
        
        truncated = et.SubElement(object_elem, 'truncated')
        truncated.text = '0'
        
        difficult = et.SubElement(object_elem, 'difficult')
        difficult.text = '0'
        
        bndbox = et.SubElement(object_elem, 'bndbox')
        xmin = et.SubElement(bndbox, 'xmin')
        xmin.text = str(x1)
        ymin = et.SubElement(bndbox, 'ymin')
        ymin.text = str(y1)
        xmax = et.SubElement(bndbox, 'xmax')
        xmax.text = str(x2)
        ymax = et.SubElement(bndbox, 'ymax')
        ymax.text = str(y2)
    out = et.tostring(root, pretty_print=True, encoding='utf8') # if element attributes contains some non ascii characters, then need to specify encoding. if not the case, then encoding doesn't need to be set
    savepath = os.path.join(savedir, f'{fn}.xml')
    with open(savepath, 'wb') as fd:
        fd.write(out)
```
